```python
#Required Imports
from feature_format import featureFormat, targetFeatureSplit
import matplotlib.pyplot as plt
plt.rcParams["patch.force_edgecolor"] = True
import seaborn as sns
sns.set_style('whitegrid')
import numpy as np
import pandas as pd
```

```python
#Loading Data
data_dict = pd.read_pickle('final_project_dataset.pkl')
```

```python
#List Of Features
features_list = ['poi','salary',
 'to_messages',
 'deferral_payments',
 'total_payments',
 'exercised_stock_options',
 'bonus',
 'restricted_stock',
 'shared_receipt_with_poi',
 'restricted_stock_deferred',
 'total_stock_value',
 'expenses',
 'loan_advances',
 'from_messages',
 'other',
 'from_this_person_to_poi',
 'director_fees',
 'deferred_income',
 'long_term_incentive',
 'from_poi_to_this_person']
```

```python
#Remove Outliers
print data_dict.pop('TOTAL',0)
print data_dict.pop('THE TRAVEL AGENCY IN THE PARK', 0)
```

```python
#Format Data
my_dataset = data_dict
data = featureFormat(my_dataset, features_list, sort_keys = True)
labels, features = targetFeatureSplit(data)
print 'Total Number Of Data Points: ',len(features)
```

```python
#Creating DataFrame From Above Data Points
enron_data = pd.DataFrame(data, columns=features_list)
enron_data.head(10)
```

```python
print "Total Number Of Poi's In The Dataset: ",sum(enron_data['poi'] == 1.0)
print "Total Number Of Non Poi's In The Dataset: ",sum(enron_data['poi'] == 0.0)
```

```python
#Salary Distribution
def draw_histogram(col,title):
    plt.figure(figsize=(10,4))
    plt.hist(enron_data[col], bins=50)
    plt.title(title);
    
draw_histogram('salary', 'Salary Distribution')
```
